```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
display(df)
```

```python
#Line chart(Trend)

plt.plot(df['Days'], df['Study_Hours'],marker='o')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
plt
```

```python
#Advance line chart
plt.figure(figsize=(7,4))

plt.plot(df['Days'], df['Study_Hours'],
         marker='o',color='green', label='Study Hours')
plt.plot(df['Days'], df['Marks'],
         marker='o',color='red', label='Marks')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
plt
```

```python
from matplotlib.lines import lineStyles
#bar chart (Comparison)
plt.bar(df['Days'], df['Marks'],color='green',edgecolor='black')
plt.title('mrs comparison')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
plt
```

```python
#advance bar graph

bars = plt.bar(df['Days'], df['Marks'],color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,y, str(y) ,ha='center',va='bottom')
plt.title('marks comparison')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.grid(axis='y',linestyle='--',alpha=0.5) #ads horizontal grid for readability
plt.show()
plt
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.arange(len(df['Days']))
width = 0.36

plt.bar(x - width/2, df['Study_Hours'],width,label='Study Hours')
plt.bar(x + width/2, df['Marks'],width,label='Marks')

plt.xticks(x,df['Days'])
plt.title('Rudansh')
plt.xlabel('Days')
plt.legend()
plt.show()
```
